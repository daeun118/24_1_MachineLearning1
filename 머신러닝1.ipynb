```python
# 딕셔너리 추출
data_dict = data.item()

# 데이터 추출
input_data = data_dict['input']
labels = data_dict['label']

# 데이터의 형태 확인
print("Input 데이터의 형태:", input_data.shape)
print("Label 데이터의 형태:", labels.shape)
```

```python
# labels 변수를 dataframe으로 변환
labels_df = pd.DataFrame(labels, columns=['label'])
labels_df
```

```python
# label 분포 시각화

colors = ["#0101DF", "#DF0101"]

sns.countplot(x='label', data=labels_df, palette=colors)
plt.title("Label Distribution")
plt.show()

# balanced data인 것을 확인함.
```
